```python
import os
import pandas as pd
import numpy as np
files = os.listdir("./Out_BGCs")
```

```python
files
```

```python
# Lập 1 từ điển có dạng {lasopeptide: {name1: 0, name2: 1}, nrps: {name1: 65, name2: 0}}
dictionary = {}
```

```python
for i, f in enumerate(files):
    # Đường dẫn hoàn chỉnh = tên thư mục + tên file nên dùng "./Out_BGCs/" + tên file
    # Ở mỗi dòng, dữ liệu cách nhau bởi 1 dấu cách, 1 dấu tab, rồi 1 dấu cách nên dùng sep = " \t "
    # Không đọc dòng đầu tiên nên dùng skip = 1
    
    dataframe = pd.read_csv("./Out_BGCs/" + f, sep = ' \t ', engine = 'python')
    data = dataframe.values
    # data có dạng 
    # [['terpene' 70]
    # ['cf_putative' 559]]
    for row in data:
        if row[0] not in dictionary:
            dictionary[row[0]] = {}
        dictionary[row[0]][f] = row[1]
```

```python
dictionary
```

```python
# Khởi tạo 1 dataframe có số dòng bằng số đối tượng, số cột bằng số file
# Tên các dòng và các cột là tên đối tượng và tên file

summary = pd.DataFrame(np.zeros((len(dictionary), len(files))), columns = files, index = dictionary.keys())
```

```python
# Xem 5 dòng đầu tiên của dataframe

summary.head(5)
```

```python
# Điền dữ liệu từ dictionary vào dataframe

for k, V in dictionary.items():
    for Vk, VV in V.items():
        summary.loc[k, Vk] = VV
```
